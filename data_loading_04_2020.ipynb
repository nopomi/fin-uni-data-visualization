```python
def get_data_from_vipunen(resource, fiql = ''):
    try:
        if len(fiql) > 0 :
            url = 'http://api.vipunen.fi/api/resources/' + resource + '/data?filter=' + fiql
        else:
            url = 'http://api.vipunen.fi/api/resources/' + resource + '/data'
        return requests.get(url, json={
            "Accept": "application/json;charset=UTF-8"
        })
    except Exception as e:
        raise e
```

```python
resp = get_data_from_vipunen('tavoiteajassa_tutkinnon_suorittaneet', 'tilastovuosi>2009 and sektori==Yliopistokoulutus')
```

```python
resp_json = json.loads(resp.text)
```
